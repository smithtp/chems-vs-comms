Some basic analyses and visualisations of our sequencing dataset. Looking at how community composition changes with increasing numbers of stressors.

Microbial communities were grown for 3 days in media supplemented with combinations of 8 chemical stressors. Optical density was measured every hour for the duration. At the final timepoint, communities were frozen at -80C. We extracted DNA and sequenced a subset of those communities to look at changes in composition. These sequenced communities were controls, each single chemical treatment, four 2-chemical treatments, 2 4-chemical treatments, and the 8-chemical treatment, such that the smaller chemical mixtures are subsets of the larger.

```r
# load some packages
library(qiime2R)
library(tidyverse)

library(ggplot2)
library(vegan)
library(dplyr)
library(lme4)
library(lmerTest)
library(ggbeeswarm)
```

```r
# load in the sequencing data (ASV table)
SVs <- read_qza("../data/Earlham/rarefied_table.qza")$data

# load in the metadata
metadata <- read.csv("../data/Earlham/metadata.tsv", sep = "\t") %>%
  rename(SampleID = sample.id)

metadata[metadata$chem.code == "",]$chem.code <- "Control"

# load in the taxonomy
taxonomy <- read_qza("../data/Earlham/taxonomy.qza")$data %>% parse_taxonomy()

# load in the phylogeny
tree <- read_qza("../data/Earlham/rooted_tree.qza")$data
```

## Taxonomy barplots

```r
# summarise the taxa to the genus level
taxasums <- summarize_taxa(SVs, taxonomy)$Genus

# plot it
options(repr.plot.width = 20, repr.plot.height = 6) # seetting plot size in jupyter
taxa_barplot(taxasums, metadata, "community.number") +
  theme_bw() +
  theme(axis.text.x = element_blank(),
        legend.text = element_text(size = 6),
        legend.position = "bottom")
```

Each of the facets shows data from a different community. Each of the bars is that community grown in a different chemical mixture (or no chemicals). Obviously doesn't tell us much directly, because I'm not showing what chemicals each community is grown in, but what it does show is that the original 10 communities clearly have different compositions.

Note that Communities 1 & 5 having less data is intentional - I had some spare wells in the 96-well plates for sequencing after putting in the other 8 communities, so stuck in a few from 1 & 5 to fill them up.

By eye, each of the communities also seems to swap between 2 different structures. I suspect this is the influence of one of our chemicals - oxytetracycline. Lets look at that:

```r
no.starting.comm <- metadata %>% filter(chem.code != "Frozen Starting Community",
                              SampleID %in% colnames(SVs))

taxa_barplot(taxasums[,no.starting.comm$SampleID], metadata, "Oxytetracycline") +
  theme_bw() +
  theme(axis.text.x = element_blank(),
        legend.text = element_text(size = 6),
        legend.position = "bottom")
```

Left facet has no oxytetracycline, right facet has oxytetracycline. Note that the ordering of communities is the same as before, so you can see the 10 communities based on their different composition and match them between the left and right panels. We're clearly losing most of the *Aeromonas* (dark green), some unassigned *Aeromonadales* (yellow), and *Janthinobacterium* (lime green), while increasing the relative abundance of *Serratia* (purple), *Enterobactereacea* (dark blue), and *Chryseobacterium* (cyan). These effects are fairly consistent across the communities despite their overall differences in composition.

This is not unexpected, given oxytetracycline is an antibiotic! It will be more interesting to know if the non-antibiotic chemicals are having a demonstrable impact on community compostition.


## Community functioning (growth)

Here we'll have a quick look at how growth of the communities was impacted by the chemicals.

```r
# create a plotting theme so we can actually read the axes
main_theme <- theme_bw() + theme(axis.text = element_text(size = 16),
                                axis.title = element_text(size = 20),
                                strip.text = element_text(size = 16))

## load in the functional data
functional_data <- read.csv("../data/spline_fits.csv")
# calculate dAUC 
control_data <- functional_data %>%
  filter(Complexity == 0) %>%
  group_by(Community, Rep) %>% # unique plate for biological replicate
  summarise(Control.AUC = mean(AUC))

# add controls to the data and calculate change
functional_dAUC <- functional_data %>%
  filter(Complexity > 0) %>%
  left_join(control_data, by = c("Community", "Rep")) %>%
  mutate(dAUC = AUC/Control.AUC)

# plot it
ggplot(functional_dAUC, aes(x = Complexity, y = dAUC)) +
  geom_beeswarm(aes(fill = as.character(Community)), shape = 21, cex = 0.5, alpha = 0.6) +
  geom_smooth(method = lm, col = "black") +
  main_theme +
  theme(legend.position = "none") +
  facet_wrap(~Community, nrow = 2)

summary(lm(dAUC ~ Community + Complexity, data = functional_dAUC))
```

Growth is measured as the area under the whole microbial growth curve (AUC). dAUC is the ratio of growth in a chemical mixture to growth under control conditions, i.e. 1 = no change, > 1 = growth increased, < 1 growth decreased.

Across all the communities, there's a slight decrease in growth as we add more chemicals. There's also a bit of bimodality in some of the community responses - those are mixtures with and without oxytetracycline again.


## Diversity Metrics

I'll have a look at diversity of the communities, and ask whether growth is linked to community diversity.

```r
# transform matrix for community phylogenetics metrics
rarefied_OTUs_t <- t(SVs)

diversity_shannon <- diversity(rarefied_OTUs_t)
diversity_simpson <- diversity(rarefied_OTUs_t, "simpson")
diversity_invsimp <- diversity(rarefied_OTUs_t, "inv")
diversity_alpha <- fisher.alpha(rarefied_OTUs_t)

species_richness <- specnumber(rarefied_OTUs_t)
evenness <- diversity_shannon/log(species_richness)

diversity_metrics <- data.frame(diversity_shannon, diversity_simpson, diversity_invsimp, diversity_alpha, species_richness,
                                evenness)
diversity_metrics$SampleID <- rownames(diversity_metrics)

# add the metadata back
diversity_metrics <- left_join(diversity_metrics, metadata)

# do some plotting
ggplot(diversity_metrics, aes(x = complexity, y = species_richness)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm) +
  labs(x = "Number of chemicals", y = "Species Richness", fill = "Community") +
  main_theme

summary(lm(scale(species_richness) ~ community.name + complexity, data = diversity_metrics))

ggplot(diversity_metrics, aes(x = complexity, y = diversity_shannon)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm) +
  labs(x = "Number of chemicals", y = "Shannon Diversity", fill = "Community") +
  main_theme

summary(lm(scale(diversity_shannon) ~ community.name + complexity, data = diversity_metrics))
```

Scaled y-variables should mean we can directly compare the strength of effects. Species richness and Shannon diversity both decrease as we add more chemicals, but the Shannon diversity is decreased by comparatively more. 

We can use the PICRUSt outputs to look at whether predicted enzymes and enzymatic pathways are also reduced along with species diversity.

```r
# calculate the same for the PICRUSt data

### Enzymes first ###

# read in the picrust data
library(data.table)
# abundance of ECs in metagenome
EC_abundance <- as.data.frame(fread("../data/Earlham/picrust/EC_metagenome_out/pred_metagenome_unstrat_descrip.tsv.gz"))

# transform these data into the sort of matrix that vegan likes
EC_matrix <- data.matrix(EC_abundance[,3:376])
dimnames(EC_matrix)[[1]] <- EC_abundance$description
EC_matrix_t <- t(EC_matrix)

# diversity metrics
EC_richness <- specnumber(EC_matrix_t)
EC_shannon <- diversity(EC_matrix_t)
EC_simpson <- diversity(EC_matrix_t, "simpson")
EC_evenness <- EC_shannon/log(EC_richness)

EC_diversity_metrics <- data.frame(EC_richness, EC_shannon, EC_simpson, EC_evenness)
EC_diversity_metrics$SampleID <- rownames(EC_diversity_metrics)

# add the metadata back
EC_diversity_metrics <- left_join(EC_diversity_metrics, metadata) %>%
  filter(chem.code != "Frozen Starting Community")

ggplot(EC_diversity_metrics, aes(x = complexity, y = EC_richness)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Enzyme Richness") +
  main_theme
  theme(legend.position = "none")

summary(lm(scale(EC_richness) ~ community.name + complexity, data = EC_diversity_metrics))

ggplot(EC_diversity_metrics, aes(x = complexity, y = EC_shannon)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Enzyme Shannon Diversity") +
  main_theme +
  theme(legend.position = "none")

summary(lm(scale(EC_shannon) ~ community.name + complexity, data = EC_diversity_metrics))
```

There's much less impact on the diversity of enzyme classes present.

```r
### now MetaCyc pathways ###
# abundance of MetaCyc pathways in metagenome
MC_abundance <- as.data.frame(fread("../data/Earlham/picrust/pathways_out/path_abun_unstrat_descrip.tsv.gz"))

MC_matrix <- data.matrix(MC_abundance[,3:375])
dimnames(MC_matrix)[[1]] <- MC_abundance$description
MC_matrix_t <- t(MC_matrix)

MC_richness <- specnumber(MC_matrix_t)
MC_shannon <- diversity(MC_matrix_t)
MC_simpson <- diversity(MC_matrix_t, "simpson")

MC_evenness <- MC_shannon/log(MC_richness)

MC_diversity_metrics <- data.frame(MC_richness, MC_shannon, MC_simpson, MC_evenness)
MC_diversity_metrics$SampleID <- rownames(MC_diversity_metrics)
MC_diversity_metrics <- left_join(MC_diversity_metrics, metadata) %>%
  filter(chem.code != "Frozen Starting Community")

ggplot(MC_diversity_metrics, aes(x = complexity, y = MC_richness)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Pathway Richness") +
  main_theme +
  theme(legend.position = "none")

summary(lm(scale(MC_richness) ~ community.name + complexity, data = MC_diversity_metrics))

ggplot(MC_diversity_metrics, aes(x = complexity, y = MC_shannon)) +
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Pathway Shannon Diversity") +
  main_theme +
  theme(legend.position = "none")

summary(lm(scale(MC_shannon) ~ community.name + complexity, data = MC_diversity_metrics))
```

Similarly, we lose a few MetaCyc pathways, but the diversity index of these isn't reduced by more chemicals. Can we argue from this that despite more chemicals reducing the species diversity of these communities, the enzymatic functioning of those communities is unlikely to be changed?


## Phylogenetic Diversity

OK, lets look at phylogenetic diversity of the communities.

```r
library(pez)

OTU_compcomm <- comparative.comm(tree, rarefied_OTUs_t) # this will take a little while to run

# calculate dispersion metrics
metrics_mntd <- .ses.mntd(OTU_compcomm)
metrics_mpd <- .ses.mpd(OTU_compcomm)

metrics_mntd$SampleID <- rownames(metrics_mntd)
metrics_mpd$SampleID <- rownames(metrics_mpd)

metrics <- left_join(metrics_mntd, metrics_mpd, by = "SampleID")
metrics <- left_join(metrics, metadata, by = "SampleID") %>%
  filter(chem.code != "Frozen Starting Community")

ggplot(metrics, aes(x = complexity, y = mntd.obs.z)) + 
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Mean Nearest Taxon Distance (MNTD)") +
  main_theme +
  theme(legend.position = "none")

summary(lm(scale(mntd.obs.z) ~ community.name + complexity, data = metrics))

ggplot(metrics, aes(x = complexity, y = mpd.obs.z)) + 
  geom_beeswarm(shape = 21, aes(fill = as.character(community.number))) +
  geom_smooth(method = lm, col = "black") +
  labs(x = "Number of chemicals", y = "Mean Phylogenetic Distance (MPD)") +
  main_theme +
  theme(legend.position = "none")

summary(lm(scale(mpd.obs.z) ~ community.name + complexity, data = metrics))
```

Quick community phylogenetic metrics refresher:

* MNTD = Mean nearest taxon distance
* MPD = Mean phylogenetic distance

If these metrics are negative, it suggests phylogenetic clustering - species in a given community are more closely related than expected. This can imply habitat filtering: species share traits that allow them to exist in an environment.

If they are positive, it suggests phylogenetic overdispersion - species are less closely related than expected. This can imply competition effects: only species that are not too ecologically similar an coexist.

In general, microbial communities are expected to be phylogenetically clustered (see  https://doi.org/10.1890/0012-9658(2006)87[100:PCAOIB]2.0.CO;2) and indeed we see that here (negative MPD and MNTD scores). However we also see that the communities are becoming less phylogenetically clustered as we increase the number of chemicals in mixture.

# Summary

In summary, as we add more chemicals to the mixtures:

* Growth is reduced (a bit)
* Species diversity is reduced
* Enzyme pathway diversity is largely unaffected
* Communities become less phylogenetically clustered

What does this tell us? I don't know. Maybe that we're losing taxa from across the whole phylogeny (rather than specific clades), but not the core taxa which provide the majority of community function? 

We need to look at the responses of specific taxa to specific chemicals to get at this more I think.
